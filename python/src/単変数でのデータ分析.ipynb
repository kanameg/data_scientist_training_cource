## データのカウント

データの件数（頻度）を集計してヒストグラムとして出力します。  
例えばタイタニックで客室クラスごとにどれくらいの人数が乗船していたかを可視化する場合などに使用します。  
seabornでは、countplotでデータの数をカウントして棒グラフで出力することが可能です。


```python
""" 棒グラフ 客室ランクごとの人数
"""
# seaborn
axis = sns.countplot(data=titanic, x='Pclass')
axis.set(xlabel='Pclass', ylabel='Count')
```

```python
""" 棒グラフ 客室ランクごとの人数（男女別）
"""
# seaborn
axis = sns.countplot(data=titanic, x='Pclass', hue='Sex')
axis.set(xlabel='Pclass', ylabel='Count')
```

```python
""" 横方向棒グラフ 客室ランクごとの人数
"""
# seaborn
axis = sns.countplot(data=titanic, y='Pclass')
axis.set(xlabel='Count', ylabel='Pclass')
```

## ヒストグラム
一つのデータの分布や傾向などを分析するのに使用します。  
タイタニックのデータでどれくらいの年齢の人がどれくらい乗船していかたかを視覚化する場合などに使用します。
seabornでは、histplotを使用して視覚化することができます。

```python
""" ヒストグラム 乗客の年齢分布
"""
axis = sns.histplot(titanic['Age'])
axis.set(xlabel='Age', ylabel='Count')
```

```python
grid = sns.FacetGrid(titanic, hue='Sex', height=7)
grid.map(sns.histplot, 'Age', kde=False)
grid.add_legend()
```
