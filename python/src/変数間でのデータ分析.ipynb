# 変数間でのデータ分析

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns

"""図のサイズ変更"""
sns.set(rc = {'figure.figsize':(10, 7)})
```

```python
""" Titanicデータ読み込み
"""
titanic = pd.read_csv('./input/titanic/train.csv')
print(titanic.head())
```

## 棒グラフ
棒グラフを使って、複数のデータ間の関連を大小で比較します。  
ここでは、タイタニックの乗客の客室ごとの年齢の平均を比較します。  
seabornではbarplotを使って棒グラフで表現が可能です。

```python
""" 棒グラフ 客室毎の平均年齢
"""
# seaborn
axis = sns.barplot(data=titanic, x='Pclass', y='Age')
axis.set(xlabel='Pclass', ylabel='Age Average')
```

```python
""" 棒グラフ 客室毎の平均年齢 生死別
"""
# seaborn
axis = sns.barplot(data=titanic, x='Pclass', y='Age', hue='Survived')
axis.set(xlabel='Pclass', ylabel='Age Average')
```

## 箱ひげ図
箱ひげ図はデータの位置や範囲について分析する手法。箱ひげ図ではデータの最大・最小・第一第二四分位数・中央値やその層にボリュームが多いかなどの分布を確認することが可能。

```python
""" 箱ひげ図 客室毎の年齢
"""
# seaborn
axis = sns.boxplot(data=titanic, x='Pclass', y='Age')
axis.set(xlabel='Pclass', ylabel='Age')
```

```python
""" 箱ひげ図 客室毎の年齢
"""
# seaborn
axis = sns.boxplot(data=titanic, x='Pclass', y='Age', hue='Survived')
axis.set(xlabel='Pclass', ylabel='Age')
```

## バイオリン図
箱ひげ図はデータの集中位置を視覚化することができるが、バイオリン図ではデータの分布の広がりを視覚化することができる。


```python
""" バイオリン図 客室毎の年齢
"""
# seaborn
axis = sns.violinplot(data=titanic, x='Pclass', y='Age')
axis.set(xlabel='Pclass', ylabel='Age')
```

```python
""" バイオリン図 客室毎の年齢を生死でスライシング
"""
# seaborn
axis = sns.violinplot(data=titanic, x='Pclass', y='Age', hue='Survived')
axis.set(xlabel='Pclass', ylabel='Age')
```

```python
""" バイオリン図 客室毎の年齢を生死でスライシング左右分割
"""
# seaborn
axis = sns.violinplot(data=titanic, x='Pclass', y='Age', hue='Survived', split=True)
axis.set(xlabel='Pclass', ylabel='Age')
```
